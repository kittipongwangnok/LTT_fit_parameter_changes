```python
#https://pyastronomy.readthedocs.io/en/latest/pyaslDoc/aslDoc/keplerOrbit.html#keplerorbitpyasl
from __future__ import print_function, division
import numpy as np
from PyAstronomy import pyasl
import matplotlib.pylab as plt

#Latex font
import matplotlib as mpl
from matplotlib import rc
plt.rc('text', usetex=True)
plt.rc('font', family='serif',size=16)
```

```python
# Instantiate a Keplerian elliptical orbit with
# semi-major axis of 1.3 length units,
# a period of 2 time units, eccentricity of 0.5,
# longitude of ascending node of 70 degrees, an inclination
# of 10 deg, and a periapsis argument of 110 deg.
#ke = pyasl.KeplerEllipse(1.3, 2., e=0.39, Omega=0., i=0.0, w=-78.0)
ke_a = pyasl.KeplerEllipse(a=8.19, per=9000 , e=0.39, Omega=0.0, i=0, w=-78)
```

```python
# Get a time axis
#t = np.linspace(0, 2.9, 500)
t_a = np.linspace(0, 9000, 500)

#print(t)
```

```python
# Calculate the orbit position at the given points
# in a Cartesian coordinate system.
pos_a = ke_a.xyzPos(t_a)
#print("Shape of output array: ", pos.shape)

# Calculate orbit radius as a function of the
radius_a = ke_a.radius(t_a)
#print(radius)

# Find the nodes of the orbit (Observer at -z)
ascn_a, descn_a = ke_a.xyzNodes_LOSZ()
#print(ascn)
```

```python
# Plot x and y coordinates of the orbit
#plt.subplot(2, 1, 1)
fig = plt.subplots(1, 1, sharex=True, sharey=False, figsize=(10, 5), tight_layout=True)
#plt.title("Periapsis (red diamond), Asc. node (green circle), desc. node (red circle)")
#plt.xlabel(r'East $\rightarrow$')
#plt.ylabel(r'North $\rightarrow$')

plt.xlabel('X (AU)')
plt.ylabel('Y (AU)')

plt.plot([0], [0], 'ko', markersize=9, label="DP Leo")
plt.plot(pos_a[::, 1], pos_a[::, 0], 'r')
# Point of periapsis
plt.plot([pos_a[0, 1]], [pos_a[0, 0]],'r*',markersize=9, label=r"Periapsis with $P$ = 9000 days")
# Nodes of the orbit
plt.plot([ascn_a[1]], [ascn_a[0]], 'rs', markersize=9,label=r"Asc. node with $P$ = 9000 days")
plt.plot([descn_a[1]], [descn_a[0]], 'rD', markersize=9,label=r"Desc. node with $P$ = 9000 days")

# Put a legend to the right of the current axis
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
